```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
df = pd.read_csv('./data_sets/insurance.csv')
```

```python
# plot the hexbin plot
sns.jointplot(x='age', y='bmi', data=df, kind='hex')    # kind='hex' is used to plot the hexbin plot
plt.show()
```

```python
# plot the contour plot
sns.jointplot(x='age', y='bmi', data=df, kind='kde')    # kind='kde' is used to plot the contour plot
plt.show()
```

```python
# plot the kde plot
sns.jointplot(x='age', y='bmi', data=df, kind='kde')    # kind='kde' is used to plot the kde plot
plt.show()
```

```python
# Plot the contingency table using crosstab
contingency_table = pd.crosstab(df['age'], df['bmi'])
print(contingency_table)


# Select the first 10 rows and columns
contingency_table_subset = contingency_table.iloc[:10, :10]

# Plot the contingency table in table form using matplotlib
fig, ax = plt.subplots(figsize=(10, 10))  # Set the figure size
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=contingency_table_subset.values, colLabels=contingency_table_subset.columns, rowLabels=contingency_table_subset.index, cellLoc='center', loc='center')

plt.show()
```

```python
# Print the povit table
pivot_table = pd.pivot_table(df, values='charges', index='age', columns='bmi')  
print(pivot_table)
```

```python

# Create a boxplot of df['Sex'] and df['Charges']
plt.figure(figsize=(10, 6))  # Set the figure size
sns.boxplot(x='sex', y='charges', data=df)
plt.title('Boxplot of Charges by Sex')
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.show()
```

```python
# Create a violin plot of df['sex'] and df['charges']
plt.figure(figsize=(10, 6))  # Set the figure size
sns.violinplot(x='sex', y='charges', data=df)
plt.title('Violin Plot of Charges by Sex')
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.show()
```

```python

# Create a FacetGrid with seaborn, faceting by 'sex' and 'region', and using 'smoker' for hue
g = sns.FacetGrid(df, col='sex', row='region', hue='smoker', margin_titles=True, height=4, aspect=1.5)
g.map(sns.histplot, 'charges', kde=True, bins=20)

# Add titles and labels
g.set_axis_labels('Charges', 'Count')
g.set_titles(col_template='{col_name} Sex | {row_name} Region')
g.add_legend()

plt.show()
```
